```python
import numpy as np 
import pandas as pd 
```

### 匯入csv檔

```python
df_baro = pd.DataFrame(pd.read_csv('./DataSource/barometer.csv',header=0))
df_rain = pd.DataFrame(pd.read_csv('./DataSource/day_rain.csv',header=0))
df_humi_in = pd.DataFrame(pd.read_csv('./DataSource/inside_humidity.csv',header=0))
df_temp_in = pd.DataFrame(pd.read_csv('./DataSource/inside_temperature.csv',header=0))
df_MPPT = pd.DataFrame(pd.read_csv('./DataSource/MPPT.csv',header=0))
df_humi_out = pd.DataFrame(pd.read_csv('./DataSource/outside_humidity.csv',header=0))
df_temp_out = pd.DataFrame(pd.read_csv('./DataSource/outside_temperature.csv',header=0))
df_solar = pd.DataFrame(pd.read_csv('./DataSource/solar_radiation.csv',header=0))
df_wind = pd.DataFrame(pd.read_csv('./DataSource/wind_speed.csv',header=0))
```

### 檢視資料空值

```python
print(df_baro.isnull().sum(),
df_rain.isnull().sum(),
df_humi_in.isnull().sum(),
df_temp_in.isnull().sum(),
df_MPPT.isnull().sum(),
df_humi_out.isnull().sum(),
df_temp_out.isnull().sum(),
df_solar.isnull().sum(),
df_wind.isnull().sum())
```

```python
# 去除重複值
df_baro=df_baro.drop_duplicates()
df_rain=df_rain.drop_duplicates()
df_humi_in=df_humi_in.drop_duplicates()
df_temp_in=df_temp_in.drop_duplicates()
df_MPPT=df_MPPT.drop_duplicates()
df_humi_out=df_humi_out.drop_duplicates()
df_temp_out=df_temp_out.drop_duplicates()
df_solar=df_solar.drop_duplicates()
df_wind=df_wind.drop_duplicates()
```

```python
#  處理dataframe 的 column name 多的空格
df_baro.columns=df_baro.columns.str.strip()
df_rain.columns =df_rain.columns.str.strip()
df_humi_in.columns =df_humi_in.columns.str.strip()
df_temp_in.columns =df_temp_in.columns.str.strip()
df_MPPT.columns =df_MPPT.columns.str.strip()
df_humi_out.columns =df_humi_out.columns.str.strip()
df_temp_out.columns =df_temp_out.columns.str.strip()
df_solar.columns =df_solar.columns.str.strip()
df_wind.columns =df_wind.columns.str.strip()
```

### 合併資料表

```python
# outer: 聯集
data_1 = pd.merge(df_baro,df_rain,on='time',how='outer')
data_2 = pd.merge(data_1,df_humi_in,on='time',how='outer')
data_3 = pd.merge(data_2,df_temp_in,on='time',how='outer')
data_4 = pd.merge(data_3,df_humi_out,on='time',how='outer')
data_5 = pd.merge(data_4,df_temp_out,on='time',how='outer')
data_6 = pd.merge(data_5,df_solar,on='time',how='outer')
data_7 = pd.merge(data_6,df_wind,on='time',how='outer')
data_total = pd.merge(data_7,df_MPPT,on='time',how='outer')
data_total
```

```python
data_total.isnull().sum()
```

```python
# inner: 交集
df_1 = pd.merge(df_baro,df_rain,on='time',how='inner')
df_2 = pd.merge(data_1,df_humi_in,on='time',how='inner')
df_3 = pd.merge(data_2,df_temp_in,on='time',how='inner')
df_4 = pd.merge(data_3,df_humi_out,on='time',how='inner')
df_5 = pd.merge(data_4,df_temp_out,on='time',how='inner')
df_6 = pd.merge(data_5,df_solar,on='time',how='inner')
df_7 = pd.merge(data_6,df_wind,on='time',how='inner')
df_total = pd.merge(data_7,df_MPPT,on='time',how='inner')
df_total
```

```python
df_total.isnull().sum()
```

```python
df_total.describe()
```

```python
# 去除空值
df = df_total.dropna()
df.isnull().sum()
```

```python
df.describe()
```

```python
df.to_csv('./DataSource/data_raw.csv')
```
