```python
dataset = dataset.drop(axis=1,columns='Unnamed: 0')
dataset
```

### Sort

```python
df = dataset.sort_values(by=['time'])
df
```

### change datatype

```python
df['time']
```

```python
start_datetime = df['time'].min()
end_datetime = df['time'].max()
```

```python
# 將object轉為 datetime64[ns](timestamp)，再將timestamp轉為int，為後續計算準備
df['timestamp'] = pd.to_datetime(df['time'])
df['timestamp'] = pd.to_numeric(df['timestamp'])/1000000000
df['timestamp']
```

### timestamp

```python
# df_data.to_csv('raw.csv')
start_timestamp = df['timestamp'].min()
end_timestamp = df['timestamp'].max()
```

```python
start_timestamp
```

```python
end_timestamp
```

```python
delta_seconds=600
ts_by_detla_seconds=[]
for ind in df['timestamp']:
    ts_by_detla_seconds.append(int((ind-start_timestamp)/delta_seconds)*delta_seconds+start_timestamp)
ts_by_detla_seconds
df
```

```python
df['ts_by_detla_seconds'] = ts_by_detla_seconds
df
```

```python
df_group = df.groupby('ts_by_detla_seconds').mean() # as_index=False會另給予index
df_group = df_group.drop(axis=1,columns='timestamp')
df_group
```

```python
df_group.to_csv('./DataSource/data_group.csv')
```

```python
df_group.describe()
```

### 去除明顯的量測錯誤值

```python
 '''
 氣壓(來源：http://www.metapp.org.tw/index.php/weatherknowledge/43-live/161-2009-01-23-06-06-50)
 在平地正常情況下，台灣地區大氣壓力大多在1000hPa±10~30hPa，冬天較高，夏季較低，有強烈颱風侵襲情況下氣壓會更低一點。
 溫度：暫且以-20~50
 風速(來源：https://sites.google.com/site/taiwanqixiangzhishiwang/home/feng-li-fen-ji-ding-yi-jie-shao)
 17.2以上為颱風等級，65以上為超級強烈颱風，故將範圍訂在0~65之間
'''
df_adjust = df_group[(df_group['value_baro']>970) & (df_group['value_baro']<1030)] # 剔除氣壓量測失敗
df_adjust = df_adjust[(df_adjust['value_temp_out']>5) & (df_adjust['value_temp_out']<50)] # 剔除溫度量測失敗
df_adjust = df_adjust[(df_adjust['value_solar']<3000)] # 剔除太陽能量測失敗
df_adjust = df_adjust[(df_adjust['value_wind']<65)] # 剔除風速量測失敗
```

```python
def get_data_figure(df_data, from_datatime, to_datatime, mean_seconds=360, moving_average_width=50):
    tup_xlim = (df_data['solar_radiation'].min(), df_data['solar_radiation'].max())
    tup_ylim = (df_data['MPPT'].min(), df_data['MPPT'].max())
    dateFormatter = "%Y/%m/%d %H:%M:%S"
    start_datatime = from_datatime
    end_datatime = to_datatime
#     print(datetime.datetime.strptime(start_datatime, dateFormatter).timestamp())
    start_unix_timestamp = int(datetime.datetime.strptime(start_datatime, dateFormatter).timestamp())
    end_unix_timestamp = int(datetime.datetime.strptime(end_datatime, dateFormatter).timestamp())
#     start_unix_timestamp = from_unixtime
#     end_unix_timestamp = to_unixtime
    start_datatime = datetime.datetime.utcfromtimestamp(start_unix_timestamp+8*60*60).strftime('%Y/%m/%d %H:%M:%S')
    end_datatime = datetime.datetime.utcfromtimestamp(end_unix_timestamp+8*60*60).strftime('%Y/%m/%d %H:%M:%S')
    df_data = df_data.loc[start_unix_timestamp:end_unix_timestamp] # 選擇特定時間的資料
    df_data = df_data.dropna()
    # df_data.to_csv('raw.csv')
    start_timestamp = df_data.index.min()
    end_timestamp = df_data.index.max()
    delta_seconds = mean_seconds
    ts_by_detla_seconds=[]
    for ind in df_data.index:
        ts_by_detla_seconds.append(int((ind-start_timestamp)/delta_seconds)*delta_seconds+start_timestamp)
    # print(ts_by_detla_seconds)
    df_data['ts_by_detla_seconds'] = ts_by_detla_seconds
    df_data = df_data.groupby('ts_by_detla_seconds').mean()
    df_data = df_data.dropna().rolling(window = moving_average_width).mean().dropna()
    # df_data
    # df_data['solar_radiation'].plot()
    # df_data.hist(bins=20)
    return df_data.plot('solar_radiation','MPPT', kind='scatter', title=start_datatime+' to '+end_datatime, xlim=tup_xlim, ylim=tup_ylim).get_figure()
#     fig.savefig('asdf.png')
```
